```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define the function and its gradient
f = lambda x: (x + 3)**2
grad_f = lambda x: 2 * (x + 3)
```

```python
# Gradient Descent
def gradient_descent(learning_rate, max_iterations, initial_x):
    history = []
    x = initial_x

    for _ in range(max_iterations):
        x -= learning_rate * grad_f(x)
        history.append(x)

    return x, history
```

```python
# Initial parameters
learning_rate = 0.1
max_iterations = 100
initial_x = 2

# Run gradient descent
min_x, history = gradient_descent(learning_rate, max_iterations, initial_x)
```

```python
# Plot the function and the descent path
x_values = np.linspace(-10, 5, 400)
y_values = f(x_values)

plt.plot(x_values, y_values, label='y = (x + 3)^2')
plt.scatter(history, [f(x) for x in history], c='red', marker='o', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Gradient Descent for y = (x + 3)^2')
plt.grid()
plt.show()

print(f"Local minimum found at x = {min_x}")
```
